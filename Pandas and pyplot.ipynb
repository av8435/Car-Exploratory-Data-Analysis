```python
#dropped owner transmission and fuel type
car=car.drop(['Fuel_Type'],axis=1)
car.head()
```

```python
#renaming columns 
car=car.rename(columns={"Selling_Price":"SP"})
car.head()
```

```python
#Total number of rows and columns
car.shape
```

```python
#Rows contatining duplicate rows
duplicate=car[car.duplicated()]
print("Number of duplicated rows:",duplicate.shape)
```

```python
#count the rows before removing the data
car.count()
```

```python
#drop duplicates
car=car.drop_duplicates()
car.count()
```

```python
#find the null values
print(car.isnull().sum())
```

```python
#drop missing values 
car=car.dropna()
car.count()
```

```python
#finding the outliers 
sns.boxplot(x=car['SP'])
```

```python
#histogram on car and numbers
car.Car_Name.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by name")
plt.ylabel("Number of cars")
plt.xlabel("Car Name");
```

```python
#find relationship between varaibles
numeric_columns=car.select_dtypes(include='number')
plt.figure(figsize=(20,10))
c=numeric_columns.corr()
sns.heatmap(c,cmap="BrBG", annot=True)
c
```
